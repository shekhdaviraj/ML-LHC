```python
ds4 = pd.read_csv('posts_short.csv')
ds4.head()
```

```python
#Determine the Number of Tags Per Question
noOfTags = {}
for i in ds4['tags']:
    flag = i.split('<')
    flag.pop(0)  
    print(flag)
    #count_tags_perQuestion[len(flag)] = count_tags_perQuestion.get(len(flag),0) + 1
    if len(flag) in noOfTags:
      noOfTags[len(flag)]+= 1
    else :
      noOfTags[len(flag)]= 1
print(noOfTags)
```

```python
xtag = list()
ytag = list()
for i,j in noOfTags.items() :
    xtag.append(i)
    ytag.append(j)
print(xtag)
print(ytag)
```

```python
plt.bar(xtag, ytag,0.5)
plt.title('Number of Ques')
plt.xlabel("Number of  tags")
plt.ylabel("total fc")
```
